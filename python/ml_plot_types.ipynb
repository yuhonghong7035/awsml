```python
%matplotlib inline
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

<h4>Plot type examples</h4>

Objective: visualize linear, quadratic, cubic, exponential, log, sine functions on a plot

```python
def quad_func (x):
    return 25*x**2 -80*x + 64
```

```python
def log_func(x):
    if x > 0:
        return math.log(x,10)
    else:
        return 0
```

```python
def straight_line(x):
    return 5*x + 8
```

```python
np.random.seed(5)
```

```python
df = pd.DataFrame(index=range(-200,200))
```

```python
df.shape
```

```python
df['x2'] = df.index.map(lambda x: x*x)
df['y_linear']=df.index.map(lambda x: 3*x+5)
df['y_quad_1'] = df.index.map(quad_func)
df['y_quad_2'] = df.index.map(lambda x: x*x - 2)
df['y_abs'] = df.index.map(lambda x: abs(x))
df['y_sinx'] = df.index.map(lambda x: math.sin(math.radians(x)))
```

```python
df['y_logx'] = df.index.map(log_func)
```

```python
df['y_x3'] = df.index.map(lambda x: x**3)
```

```python
df['y_2powx'] = df.index.map(lambda x: 2.0 **x)
```

```python
df.shape
```

```python
df.head()
```

```python
df.plot(y='y_linear', title='linear 5x+8', grid=True)
```

```python
df.plot(y='y_quad_1',title='Second Order 25x^2-80x+64', grid=True)
```

```python
df.plot(y='y_x3', title='Third Order x^3', grid=True)
```

```python
df.plot(y='y_2powx',title='Exponential 2^x',xlim=(0,20), ylim=(0,30000), grid=True)
```

```python
df.plot(y='y_logx', title='log(x)', xlim=((0,200)), grid=True)
```

```python
df.plot(y='y_sinx', title='sin', grid=True)
```
